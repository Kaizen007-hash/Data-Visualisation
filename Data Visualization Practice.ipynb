```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate Fibonacci sequence
def fibonacci(n):
    fib_sequence = [0, 1]
    for i in range(2, n):
        fib_sequence.append(fib_sequence[i-1] + fib_sequence[i-2])
    return fib_sequence

# Generate first 100 Fibonacci numbers
n = 100
fib_numbers = fibonacci(n)

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(range(1, n+1), fib_numbers, marker='o', linestyle='-', markersize=4)
plt.title('First 100 Fibonacci Numbers')
plt.xlabel('Position in Sequence')
plt.ylabel('Value')
plt.grid(True)
plt.yscale('log')  # Using log scale as Fibonacci numbers grow exponentially

# Add a second subplot to show the first few numbers more clearly
plt.figure(figsize=(12, 6))
plt.plot(range(1, min(21, n+1)), fib_numbers[:20], marker='o', linestyle='-', color='green')
plt.title('First 20 Fibonacci Numbers')
plt.xlabel('Position in Sequence')
plt.ylabel('Value')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
from plotly.offline import iplot
import plotly.express as px
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv('DA_1.csv')

# Apply Sigmoid function to filter T1
df['Sigmoid_T1'] = 1 / (1 + np.exp(-df['T1']))
filtered_df = df[df['Sigmoid_T1'] > 0.5]

# Extract X and Y for regression
X = filtered_df[['T2']].values  # Independent variable
Y = filtered_df['T1'].values    # Dependent variable

# Apply Polynomial Transformation (degree=3 for best fit)
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)

# Train Linear Regression Model
model = LinearRegression()
model.fit(X_poly, Y)

# Generate predictions
X_sorted = np.sort(X, axis=0)  # Sort values for smooth plotting
Y_pred = model.predict(poly.transform(X_sorted))

# Plot Scatter Plot + Polynomial Regression Curve
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X_sorted, Y_pred, color='red', linewidth=2, label='Polynomial Fit')
plt.xlabel('T2')
plt.ylabel('T1')
plt.title('Polynomial Regression (Best Fit Line)')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv('DA_1.csv')

# Data cleaning
df = df.drop_duplicates().dropna()
df['T1'] = pd.to_numeric(df['T1'], errors='coerce')
df['T2'] = pd.to_numeric(df['T2'], errors='coerce')

# Ensure dataset isn't empty
if df.empty:
    raise ValueError("Dataset is empty after cleaning. Check your data!")

# Fit Polynomial Regression (Degree 3 for curvature)
degree = 9
poly = PolynomialFeatures(degree)

X = df[['T2']].values  # Independent Variable
y = df['T1'].values  # Dependent Variable
X_poly = poly.fit_transform(X)

# Train the model with forced zero intercept
model = LinearRegression(fit_intercept=False)
model.fit(X_poly, y)

# Ensure the curve faces downward (negative leading coefficient)
if model.coef_[-1] > 0:  
    model.coef_ = -model.coef_  # Flip the sign of all coefficients

# Generate smooth curve points
X_range = np.linspace(X.min(), X.max(), 300).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
y_pred = model.predict(X_range_poly)

# Extract Polynomial Equation
coefs = model.coef_
equation = "T1 = " + " + ".join([f"{coefs[i]:.2f} * T2^{i}" for i in range(1, degree,  1)])

# Plot scatter points
fig = px.scatter(df, x='T2', y='T1', title="Downward Polynomial Regression Curve", color='T2')

# Add regression curve
fig.add_trace(go.Scatter(
    x=X_range.flatten(), y=y_pred, 
    mode='lines', name="Polynomial Fit", 
    line=dict(color='red', width=3)
))

# Display equation on plot
fig.update_layout(annotations=[
    dict(text=equation, xref="paper", yref="paper", x=0.05, y=0.90, showarrow=False, font=dict(size=14, color="white"))
])

# Show plot
fig.show()

# Print polynomial equation
print("Polynomial Regression Equation:")
print(equation)
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv('DA_1.csv')
assert 'T1' in df.columns and 'T2' in df.columns, "Columns 'T1' or 'T2' not found in dataset."

# Data cleaning
df = df.drop_duplicates().dropna()
df[['T1', 'T2']] = df[['T1', 'T2']].apply(pd.to_numeric, errors='coerce')

# Fit Polynomial Regression
degree = 9
poly = PolynomialFeatures(degree)
X = df[['T2']].values  # Ensure X is 2D
y = df['T1'].values  # Dependent Variable
X_poly = poly.fit_transform(X)

# Train the model
model = LinearRegression()
model.fit(X_poly, y)

# Generate predictions for smooth curve
X_range = np.linspace(X.min(), X.max(), 300).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
y_pred = model.predict(X_range_poly)

# Extract Polynomial Equation
coefs = model.coef_
intercept = model.intercept_
equation = f"T1 = {intercept:.2f} " + " + ".join([f"{coefs[i]:.2f} * T2^{i}" for i in range(1, degree + 1)])

# Plot scatter points and curve
fig = go.Figure()
fig.add_trace(go.Scatter(x=df['T2'], y=df['T1'], mode='markers', name="Data Points"))
fig.add_trace(go.Scatter(x=X_range.flatten(), y=y_pred, mode='lines', name="Polynomial Fit", line=dict(color='red', width=3)))

# Add annotations
fig.update_layout(
    title="Polynomial Regression Curve",
    xaxis_title="T2",
    yaxis_title="T1",
    annotations=[dict(text=equation, x=0.5, y=-0.2, xref="paper", yref="paper", showarrow=False, font=dict(size=14, color="black"))]
)

# Show plot
fig.show()

# Print polynomial equation
print("Polynomial Regression Equation:")
print(equation)
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv('DA_1.csv')
assert 'T1' in df.columns and 'T2' in df.columns, "Columns 'T1' or 'T2' not found in dataset."

# Data cleaning
df = df.drop_duplicates().dropna()
df[['T1', 'T2']] = df[['T1', 'T2']].apply(pd.to_numeric, errors='coerce')

# Flip the target variable (T1) for a downward curve
degree = 9
poly = PolynomialFeatures(degree)
X = df[['T2']].values  # Ensure X is 2D
y = -df['T1'].values  # Flip target variable

# Fit Polynomial Regression
X_poly = poly.fit_transform(X)
model = LinearRegression()
model.fit(X_poly, y)

# Generate predictions for smooth curve
X_range = np.linspace(X.min(), X.max(), 300).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
y_pred = -model.predict(X_range_poly)  # Invert predictions to restore scale

# Extract Polynomial Equation
coefs = -model.coef_  # Invert coefficients to reflect the original curve
intercept = -model.intercept_  # Invert the intercept
equation = f"T1 = {intercept:.2f} " + " + ".join([f"{coefs[i]:.2f} * T2^{i}" for i in range(1, degree + 1)])

# Plot scatter points and curve
fig = go.Figure()
fig.add_trace(go.Scatter(x=df['T2'], y=df['T1'], mode='markers', name="Data Points"))
fig.add_trace(go.Scatter(x=X_range.flatten(), y=y_pred, mode='lines', name="Polynomial Fit", line=dict(color='red', width=3)))

# Add annotations
fig.update_layout(
    title="Polynomial Regression Curve (Downward Facing)",
    xaxis_title="T2",
    yaxis_title="T1",
    annotations=[dict(text=equation, x=0.5, y=-0.2, xref="paper", yref="paper", showarrow=False, font=dict(size=14, color="black"))]
)

# Show plot
fig.show()

# Print polynomial equation
print("Polynomial Regression Equation (Downward Facing):")
print(equation)
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Load and clean dataset
df = pd.read_csv('DA_1.csv').drop_duplicates().dropna()
df[['T1', 'T2']] = df[['T1', 'T2']].apply(pd.to_numeric, errors='coerce')
df = df.dropna(subset=['T1', 'T2'])

# Polynomial Regression
degree = 9
poly = PolynomialFeatures(degree)
X = df[['T2']].values
y = df['T1'].values
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)
df['T1_poly'] = model.predict(X_poly)

# Sigmoid function to normalize fitted values
def sigmoid(x, mu, sigma):
    sigma = sigma if sigma != 0 else 1
    return 1 / (1 + np.exp(-(x - mu) / sigma))

mu = df['T1_poly'].mean()
sigma = df['T1_poly'].std()
df['T1_sigmoid'] = sigmoid(df['T1_poly'], mu, sigma)

fig = px.scatter(df, x='T2', y='T1_sigmoid', color='T2', title="Sigmoid Filtered Data")

X_range = np.linspace(X.min(), X.max(), 300).reshape(-1, 1)
y_pred = model.predict(poly.transform(X_range))
y_pred_sigmoid = sigmoid(y_pred, mu, sigma)

fig.add_trace(go.Scatter(x=X_range.flatten(), y=y_pred_sigmoid, mode='lines', name="Sigmoid Fit", line=dict(color='red', width=3)))
fig.show()

coeffs = model.coef_
equation_terms = [f"{model.intercept_:.3f}"] + [f"{coeffs[i]:.3f}x^{i}" for i in range(1, degree + 1)]
equation = " + ".join(equation_terms)
print(f"Polynomial Equation: T1 = {equation}")
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from scipy.stats import pearsonr

# TensorFlow is not usable in this Python 3.13 environment, so this cell uses
# a scikit-learn neural network on rolling time-window features.
df = pd.read_csv('DA_1.csv').drop_duplicates().dropna()
df[['T1', 'T2']] = df[['T1', 'T2']].apply(pd.to_numeric, errors='coerce')
df = df.dropna(subset=['T1', 'T2']).reset_index(drop=True)

sequence_length = 10
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[['T1', 'T2']])

X = np.array([scaled[i:i + sequence_length].ravel() for i in range(len(scaled) - sequence_length)])
y = scaled[sequence_length:, 0]

model = MLPRegressor(hidden_layer_sizes=(64, 32), activation='relu', max_iter=500, random_state=42)
model.fit(X, y)

pred_scaled = model.predict(X).reshape(-1, 1)
predictions = scaler.inverse_transform(np.column_stack((pred_scaled, np.zeros_like(pred_scaled))))[:, 0]
t1_inverse = scaler.inverse_transform(scaled)[:, 0]
actual_t1_aligned = t1_inverse[sequence_length:]

fig = go.Figure()
fig.add_trace(go.Scatter(y=actual_t1_aligned, mode='lines', name="Actual T1"))
fig.add_trace(go.Scatter(y=predictions, mode='lines', name="MLP Predicted T1", line=dict(color='red')))
fig.update_layout(title="T1 Forecast Using Rolling Windows", xaxis_title="Time", yaxis_title="T1", template="plotly_dark")
fig.show()

mse = mean_squared_error(actual_t1_aligned, predictions)
mae = mean_absolute_error(actual_t1_aligned, predictions)
r2 = r2_score(actual_t1_aligned, predictions)
correlation, p_value = pearsonr(actual_t1_aligned, predictions)
print(f"Mean Squared Error (MSE): {mse:.6f}")
print(f"Mean Absolute Error (MAE): {mae:.6f}")
print(f"R-squared (R2): {r2:.4f}")
print(f"Pearson correlation: {correlation:.4f} (p={p_value:.4e})")
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from scipy.stats import pearsonr

# TensorFlow is not usable in this Python 3.13 environment, so this cell uses
# a scikit-learn neural network on rolling time-window features.
df = pd.read_csv('DA_1.csv').drop_duplicates().dropna()
df[['T1', 'T2']] = df[['T1', 'T2']].apply(pd.to_numeric, errors='coerce')
df = df.dropna(subset=['T1', 'T2']).reset_index(drop=True)

sequence_length = 10
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[['T1', 'T2']])

X = np.array([scaled[i:i + sequence_length].ravel() for i in range(len(scaled) - sequence_length)])
y = scaled[sequence_length:, 0]

model = MLPRegressor(hidden_layer_sizes=(64, 32), activation='relu', max_iter=500, random_state=42)
model.fit(X, y)

pred_scaled = model.predict(X).reshape(-1, 1)
predictions = scaler.inverse_transform(np.column_stack((pred_scaled, np.zeros_like(pred_scaled))))[:, 0]
t1_inverse = scaler.inverse_transform(scaled)[:, 0]
actual_t1_aligned = t1_inverse[sequence_length:]

fig = go.Figure()
fig.add_trace(go.Scatter(y=actual_t1_aligned, mode='lines', name="Actual T1"))
fig.add_trace(go.Scatter(y=predictions, mode='lines', name="MLP Predicted T1", line=dict(color='red')))
fig.update_layout(title="T1 Forecast Using Rolling Windows", xaxis_title="Time", yaxis_title="T1", template="plotly_dark")
fig.show()

mse = mean_squared_error(actual_t1_aligned, predictions)
mae = mean_absolute_error(actual_t1_aligned, predictions)
r2 = r2_score(actual_t1_aligned, predictions)
correlation, p_value = pearsonr(actual_t1_aligned, predictions)
print(f"Mean Squared Error (MSE): {mse:.6f}")
print(f"Mean Absolute Error (MAE): {mae:.6f}")
print(f"R-squared (R2): {r2:.4f}")
print(f"Pearson correlation: {correlation:.4f} (p={p_value:.4e})")
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from pykalman import KalmanFilter

# Load dataset
df = pd.read_csv('DA_1.csv')

# Data Cleaning
df = df.drop_duplicates().dropna()

# Ensure T1 and T2 are numeric
df['T1'] = pd.to_numeric(df['T1'], errors='coerce')
df['T2'] = pd.to_numeric(df['T2'], errors='coerce')

# Drop rows with NaN after conversion
df = df.dropna(subset=['T1', 'T2'])

# Apply Kalman Filter to smooth T1
kf = KalmanFilter(initial_state_mean=0, n_dim_obs=1)
kalman_smoothed, _ = kf.em(df['T1'].values).smooth(df['T1'].values)

# Store filtered T1 in DataFrame
df['T1_filtered'] = kalman_smoothed.flatten()

# Apply Polynomial Regression on filtered T1
degree = 9  # You can adjust this
coeffs = np.polyfit(df['T2'], df['T1_filtered'], degree)
poly_eq = np.poly1d(coeffs)

# Generate fitted values
df['T1_pred'] = poly_eq(df['T2'])

# Scatter Plot: Actual Filtered Data
fig = px.scatter(df, x='T2', y='T1_filtered', color='T2',
                 template='plotly_dark', title="Kalman-Filtered T1 vs. T2 with Polynomial Fit")

# Regression Line
fig.add_trace(go.Scatter(
    x=df['T2'], 
    y=df['T1_pred'], 
    mode='lines', 
    line=dict(color='red', width=3),
    name='Polynomial Fit on Filtered T1'
))

# Show plot
fig.show()

# Print Polynomial Equation
equation = " + ".join([
    f"{coeff:.3f}x^{i}" if i > 0 else f"{coeff:.3f}" 
    for i, coeff in enumerate(coeffs[::-1])
])
print(f"Polynomial Equation: T1_filtered = {equation}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import MinMaxScaler

# Residual diagnostics for the scikit-learn rolling-window forecast.
df = pd.read_csv('DA_1.csv').drop_duplicates().dropna()
df[['T1', 'T2']] = df[['T1', 'T2']].apply(pd.to_numeric, errors='coerce')
df = df.dropna(subset=['T1', 'T2']).reset_index(drop=True)

sequence_length = 10
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[['T1', 'T2']])

X = np.array([scaled[i:i + sequence_length].ravel() for i in range(len(scaled) - sequence_length)])
y = scaled[sequence_length:, 0]

model = MLPRegressor(hidden_layer_sizes=(64, 32), activation='relu', max_iter=500, random_state=42)
model.fit(X, y)

pred_scaled = model.predict(X).reshape(-1, 1)
df_pred = scaler.inverse_transform(np.column_stack((pred_scaled, np.zeros_like(pred_scaled))))[:, 0]
t1_inverse = scaler.inverse_transform(scaled)[:, 0]
actual_t1_aligned = t1_inverse[sequence_length:]

plt.figure(figsize=(14, 6))
plt.plot(actual_t1_aligned, label='Actual T1', color='blue')
plt.plot(df_pred, label='MLP Predicted T1', color='red', alpha=0.7)
plt.title("Actual vs. Predicted T1")
plt.xlabel("Time Steps")
plt.ylabel("T1 Value")
plt.legend()
plt.grid(True)
plt.show()

residuals = actual_t1_aligned - df_pred
plt.figure(figsize=(10, 5))
plt.plot(residuals, label='Residuals', color='purple')
plt.axhline(y=0, color='gray', linestyle='--')
plt.title("Residuals (Actual - Predicted)")
plt.xlabel("Time Steps")
plt.ylabel("Error")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(8, 4))
plt.hist(residuals, bins=30, color='orange', edgecolor='black')
plt.title("Distribution of Residuals")
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Animate the latest actual/predicted series if the previous forecasting cell
# has been run; otherwise rebuild a small forecast so this cell is self-contained.
if 'actual_t1_aligned' not in globals() or 'df_pred' not in globals():
    import numpy as np
    from sklearn.neural_network import MLPRegressor
    from sklearn.preprocessing import MinMaxScaler

    df = pd.read_csv('DA_1.csv').drop_duplicates().dropna()
    df[['T1', 'T2']] = df[['T1', 'T2']].apply(pd.to_numeric, errors='coerce')
    df = df.dropna(subset=['T1', 'T2']).reset_index(drop=True)
    sequence_length = 10
    scaler = MinMaxScaler()
    scaled = scaler.fit_transform(df[['T1', 'T2']])
    X = np.array([scaled[i:i + sequence_length].ravel() for i in range(len(scaled) - sequence_length)])
    y = scaled[sequence_length:, 0]
    model = MLPRegressor(hidden_layer_sizes=(64, 32), max_iter=500, random_state=42)
    model.fit(X, y)
    pred_scaled = model.predict(X).reshape(-1, 1)
    df_pred = scaler.inverse_transform(np.column_stack((pred_scaled, np.zeros_like(pred_scaled))))[:, 0]
    actual_t1_aligned = scaler.inverse_transform(scaled)[:, 0][sequence_length:]

df_anim = pd.DataFrame({
    'Time': range(len(df_pred)),
    'Actual_T1': actual_t1_aligned[:len(df_pred)],
    'Predicted_T1': df_pred
})

df_melted = df_anim.melt(id_vars='Time', value_vars=['Actual_T1', 'Predicted_T1'], var_name='Series', value_name='T1_Value')

fig = px.line(
    df_melted,
    x='Time',
    y='T1_Value',
    color='Series',
    animation_frame='Time',
    title='Animated Time Series: Actual vs Predicted T1',
    range_y=[df_melted['T1_Value'].min(), df_melted['T1_Value'].max()],
    template='plotly_dark'
)
fig.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neural_network import MLPRegressor

# Simple prediction engine: predict T1 from T2 with scikit-learn.
df = pd.read_csv('DA_1.csv').drop_duplicates().dropna()
df[['T1', 'T2']] = df[['T1', 'T2']].apply(pd.to_numeric, errors='coerce')
df = df.dropna(subset=['T1', 'T2'])

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[['T2']])
target = df['T1'].values

X_train, X_test, y_train, y_test = train_test_split(scaled_data, target, test_size=0.2, random_state=42)

model = MLPRegressor(hidden_layer_sizes=(64, 32), activation='relu', max_iter=500, random_state=42)
model.fit(X_train, y_train)
predictions = model.predict(X_test)

for i in range(min(5, len(predictions))):
    print(f"Predicted T1: {predictions[i]:.2f} | Actual T1: {y_test[i]:.2f}")
```
